# **TRANSFORMASI LINIER**

### **Pengertian :**
Transformasi adalah pemetaan T dari ruang vektor V ke ruang vektor W.

Cara menentukan tranformasi linier :
- pusat titik tidak berubah
- garis lurus ditetapkan sebagai lurus

### **Secara umum pemetaan $T$ adalah transformasi Linier jika**
- $T(u+v)=T(u)+T(v)$
- $T (αv) = αT (v)$

#### Pembuktian :  

fungsi $ T: \mathbb{R}^2 \to \mathbb{R}^2 $ dengan $ T(x, 2y) = (-x, y) $

##### - Sifat 1 : $T(u+v)=T(u)+T(v)$

1. Misalkan $ u = (x_1, 2y_1) $ dan $ v = (x_2, 2y_2) $ adalah dua vektor di $ \mathbb{R}^2 $.

2. Hitung $ T(u) $:   
   
   $
   T(u) = T(x_1, 2y_1) = (-x_1, y_1)
   $

3. Hitung $ T(v) $:   
   
   $
   T(v) = T(x_2, 2y_2) = (-x_2, y_2)
   $

4. Hitung $ u + v $:   
   
   $
   u + v = (x_1, 2y_1) + (x_2, 2y_2) = (x_1 + x_2, 2(y_1 + y_2))
   $

5. Terapkan $ T $ pada $ u + v $:   
   
   $
   T(u + v) = T(x_1 + x_2, 2(y_1 + y_2)) = (-(x_1 + x_2), y_1 + y_2)
   $

6. Hitung $ T(u) + T(v) $:   
   
   $
   T(u) + T(v) = (-x_1, y_1) + (-x_2, y_2) = (-x_1 - x_2, y_1 + y_2)
   $

7. Bandingkan $ T(u + v) $ dan $ T(u) + T(v) $:   
   
   $
   T(u + v) = (-(x_1 + x_2), y_1 + y_2)
   $
   $
   T(u) + T(v) = (-x_1 - x_2, y_1 + y_2)
   $

Karena $ T(u + v) = T(u) + T(v) $, sifat 1 terpenuhi.

#### - Sifat 2 : $T (αv) = αT (v)$

1. Misalkan $ v = (x, 2y) $ adalah vektor di $ \mathbb{R}^2 $ dan $ \alpha $ adalah skalar.

2. Hitung $ T(v) $:   
   
   $
   T(v) = T(x, 2y) = (-x, y)
   $

3. Hitung $ \alpha v $:   
   
   $
   \alpha v = \alpha (x, 2y) = (\alpha x, 2\alpha y)
   $

4. Terapkan $ T $ pada $ \alpha v $:   
   
   $
   T(\alpha v) = T(\alpha x, 2\alpha y) = (-\alpha x, \alpha y)
   $

5. Hitung $ \alpha T(v) $:   
   
   $
   \alpha T(v) = \alpha (-x, y) = (-\alpha x, \alpha y)
   $

6. Bandingkan $ T(\alpha v) $ dan $ \alpha T(v) $:   
   
   $
   T(\alpha v) = (-\alpha x, \alpha y)
   $
   $
   \alpha T(v) = (-\alpha x, \alpha y)
   $

Karena $ T(\alpha v) = \alpha T(v) $, sifat 2 terpenuhi.

### **Menghitung matriks A dari Transformasi matriks berikut :**

$$
T\left[\begin{array}{l}
2 \\
0
\end{array}\right]=\left[\begin{array}{l}
-2 \\
0
\end{array}\right], T\left[\begin{array}{r}
2 \\
2
\end{array}\right]=\left[\begin{array}{l}
-2 \\
2
\end{array}\right]
$$


#### Langkah-langkah:

1. Nyatakan vektor basis standar:   
   Vektor standar basis untuk $\mathbb{R}^2$ adalah $\mathbf{e}_1 = \begin{bmatrix}1 \\ 0\end{bmatrix}$ dan $\mathbf{e}_2 = \begin{bmatrix}0 \\ 1\end{bmatrix}$.

2. Gunakan vektor-vektor yang diberikan untuk membentuk sistem persamaan:   
   Vektor $ \begin{bmatrix} 2 \\ 0 \end{bmatrix} $ dapat dinyatakan sebagai $ 2 \mathbf{e}_1 $.   
   Vektor $ \begin{bmatrix} 2 \\ 2 \end{bmatrix} $ dapat dinyatakan sebagai $ 2 \mathbf{e}_1 + 2 \mathbf{e}_2 $.

3. Ekspresikan hasil transformasi:   
   
   $
   T\left( 2 \mathbf{e}_1 \right) = \begin{bmatrix} -2 \\ 0 \end{bmatrix} \implies T(\mathbf{e}_1) = \begin{bmatrix} -1 \\ 0 \end{bmatrix}
   $   
   
   $
   T\left( 2 \mathbf{e}_1 + 2 \mathbf{e}_2 \right) = \begin{bmatrix} -2 \\ 2 \end{bmatrix} \implies T(\mathbf{e}_1) + T(\mathbf{e}_2) = \begin{bmatrix} -1 \\ 0 \end{bmatrix} + T(\mathbf{e}_2) = \begin{bmatrix} -1 \\ 0 \end{bmatrix} + T(\mathbf{e}_2) = \begin{bmatrix} -1 \\ 0 \end{bmatrix} + \begin{bmatrix} 0 \\ 2 \end{bmatrix}
   $   
   
   Dari sini kita lihat bahwa:   
   
   $
   T(\mathbf{e}_2) = \begin{bmatrix} 0 \\ 1 \end{bmatrix}
   $

4. Konstruksi matriks transformasi $ A $:   
   Matriks $ A $ yang merepresentasikan $ T $ dalam basis standar adalah matriks yang kolom-kolomnya adalah hasil transformasi dari vektor-vektor basis standar.   
   
   $
   A = \begin{bmatrix} T(\mathbf{e}_1) & T(\mathbf{e}_2) \end{bmatrix} = \begin{bmatrix} -1 & 0 \\ 0 & 1 \end{bmatrix}
   $

Jadi, matriks $ A $ dari transformasi linier $ T $ adalah:
$
A = \begin{bmatrix} -1 & 0 \\ 0 & 1 \end{bmatrix}
$

### **Refleksi Matriks**
Tranformasi Linier dengan Refleksi atau pencerminan pada tranformasi linear adalah adalah salah satu metode tranformasi linear yang digunakan untuk mencerminkan suatu Node dengan fungsi matrix. Berikut adalah sifat-sifat Refleksi:

1. Objek yang direfleksikan tidak mengalami perubahan benntuk dan ukuran

2. Jarak objek dari cermin (sumbu X/Y) adalah sama dengan jarak bayangan dengan cermin tersbut.

Contoh :   
Misalkan refleksi dari titik $ A(1, 1)$ dan bentuk matriksnya yaitu :

$
A \begin{bmatrix} 
1 & 2 \\
2 & 1 
\end{bmatrix}
$

- Refleksi terhadap sumbu $ x $:  

   $
   A_x = \begin{bmatrix} 1 & -2 \\ -2 & 1 \end{bmatrix}
   $

   Titik $ A(1, 1) $ menjadi $ A(1, -1) $.

- Refleksi terhadap sumbu $ y $:  

   $
   A_y = \begin{bmatrix} 1 & -2 \\ 2 & 1 \end{bmatrix}
   $  

   Titik $ A(1, 1) $ menjadi $ A(-1, 1) $.

- Refleksi terhadap garis $ x = y $:  

   $
   A_{x=y} = \begin{bmatrix} 1 & 2 \\ 2 & 1 \end{bmatrix}
   $  
   
   Titik $ A(1, 1) $ tetap $ A(1, 1) $.
